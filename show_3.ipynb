```python
# From binary to visuals for consumption, 2
# The scatter plot availed little.
# Perhaps a line graph, with added explanation. 
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = np.random.randint(0, 101, 500)
x = np.random.randint(0, 11, 500)
```

```python
plt.plot(x, y, c = "magenta")
plt.ylabel("Out of 100")
plt.xlabel("Out of 10")
plt.title("Random Distribution 500", size=18)
plt.show()
```

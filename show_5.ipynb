```python
# Import 'python libraries'
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create 500 random numbers between 0 and 100,
## that will be associated with a number 
## between 0 and 10
y = np.random.randint(0, 101, 500)
x = np.random.randint(0, 11, 500)
y.sort()
x.sort()
```

```python
plt.style.use("fivethirtyeight")
plt.hist(y, bins=25, edgecolor="white")
plt.ylabel("Out of 100")
plt.xlabel("Out of 10")
plt.title("Random Distribution 500", size=18)
plt.show()
```
